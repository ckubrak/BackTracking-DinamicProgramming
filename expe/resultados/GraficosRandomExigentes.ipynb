```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DinRandom = pd.read_csv('DinExigenteRandom.csv1.txt', delim_whitespace=True)
BackFactRandom = pd.read_csv('BackExigenteRandom.csv1.txt', delim_whitespace=True)

DinRandom2 = pd.read_csv('DinExigenteRandom.csv2.txt', delim_whitespace=True)
BackFactRandom2 = pd.read_csv('BackExigenteRandom.csv2.txt', delim_whitespace=True)

DinRandom3 = pd.read_csv('DinExigenteRandom.csv3.txt', delim_whitespace=True)
BackFactRandom3 = pd.read_csv('BackExigenteRandom.csv3.txt', delim_whitespace=True)

general = DinRandom

general['tiempoBack'] = (BackFactRandom['tiempo'] + BackFactRandom2['tiempo'] + BackFactRandom['tiempo'])/3
general['llamadasBack'] = (BackFactRandom['llamadas'] + BackFactRandom2['llamadas'] + BackFactRandom['llamadas'])//3

print(general)
```

# V fijo

```python
vFijoChico = general.loc[general['v'] == 25];
vFijoGrande = general.loc[general['v'] == 5929];
```

###### Graficos de llamadas

```python
plot = sns.lineplot(x="n", y="llamadas", data=vFijoChico);
plot = sns.lineplot(x="n", y="llamadasBack", data=vFijoChico);


plot.legend(["Dinamica", "BackTracking"]);

plot.set(ylabel="Llamadas recursivas");
```

```python
plot = sns.lineplot(x="n", y="llamadasDin", data=vFijoChico,estimator=None);
plot = sns.lineplot(x="n", y="llamadasBack", data=vFijoChico,estimator=None);


plot.legend(["Dinamica", "BackTracking"]);

plot.set(ylabel="Llamadas recursivas");
```

```python
plot = sns.lineplot(x="n", y="llamadas", data=vFijoGrande,estimator=None);
plot = sns.lineplot(x="n", y="llamadasBack", data=vFijoGrande,estimator=None);


plot.legend(["Dinamica", "BackTracking"]);

plot.set(ylabel="Llamadas recursivas");
```

###### Graficos Tiempo

```python
plot = sns.lineplot(x="n", y="tiempo", data=vFijoChico);
plot = sns.lineplot(x="n", y="tiempoBack", data=vFijoChico);


plot.legend(["Dinamica", "BackTracking"]);
#plot.legend(["BackTracking Factibilidad"]);

plot.set(ylabel="Tiempo (ms)");
```

```python
plot = sns.lineplot(x="n", y="tiempo", data=vFijoChico,estimator=None);
plot = sns.lineplot(x="n", y="tiempoBack", data=vFijoChico,estimator=None);


plot.legend(["Dinamica", "BackTracking"]);
#plot.legend(["BackTracking Factibilidad"]);

plot.set(ylabel="Tiempo (ms)");
```

```python
plot = sns.lineplot(x="n", y="tiempoDin", data=vFijoGrande,estimator=None);
plot = sns.lineplot(x="n", y="tiempoBack", data=vFijoGrande,estimator=None);


plot.legend(["Dinamica", "BackTracking"]);
#plot.legend(["BackTracking Factibilidad"]);

plot.set(ylabel="Tiempo (ms)");
```
