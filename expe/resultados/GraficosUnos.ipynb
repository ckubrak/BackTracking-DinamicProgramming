```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
FBRandom = pd.read_csv('FBRandom.csv1.txt', delim_whitespace=True)
DinRandom = pd.read_csv('DinRandom.csv1.txt', delim_whitespace=True)
BackFactRandom = pd.read_csv('BackRandom.csv1.txt', delim_whitespace=True)
general = FBRandom

general['tiempoDin'] = DinRandom['tiempo']
general['llamadasDin'] = DinRandom['llamadas'] 

general['tiempoBack'] = BackFactRandom['tiempo'] 
general['llamadasBack'] = BackFactRandom['llamadas']

print(general)
```

# V fijo

```python
vFijoChico = general.loc[general['v'] == 16];
vFijoGrande = general.loc[general['v'] == 6241];
```

###### Graficos de llamadas

```python
plot = sns.lineplot(x="n", y="llamadas", data=vFijoChico);
plot = sns.lineplot(x="n", y="llamadasDin", data=vFijoChico);
plot = sns.lineplot(x="n", y="llamadasBack", data=vFijoChico);


plot.legend(["Fuerza Bruta", "Dinamica", "BackTracking"]);

plot.set(ylabel="Llamadas recursivas");
```

```python
plot = sns.lineplot(x="n", y="llamadasDin", data=vFijoChico,estimator=None);
plot = sns.lineplot(x="n", y="llamadasBack", data=vFijoChico,estimator=None);


plot.legend(["Dinamica", "BackTracking"]);

plot.set(ylabel="Llamadas recursivas");
```

```python
plot = sns.lineplot(x="n", y="llamadasDin", data=vFijoGrande,estimator=None);
plot = sns.lineplot(x="n", y="llamadasBack", data=vFijoGrande,estimator=None);


plot.legend(["Dinamica", "BackTracking"]);

plot.set(ylabel="Llamadas recursivas");
```

###### Graficos Tiempo

```python
plot = sns.lineplot(x="n", y="tiempo", data=vFijoChico);
plot = sns.lineplot(x="n", y="tiempoDin", data=vFijoChico);
plot = sns.lineplot(x="n", y="tiempoBack", data=vFijoChico);


plot.legend(["Fuerza Bruta", "Dinamica", "BackTracking"]);
#plot.legend(["BackTracking Factibilidad"]);

plot.set(ylabel="Tiempo (ms)");
```
